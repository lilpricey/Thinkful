# Automobiles

The automobiles dataset contains information on various types of cars. It's from the UCI Machine Learning Repository and more information can be found about it [Here](http://archive.ics.uci.edu/ml/datasets/Automobile).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('Automobile.csv')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
sns.distplot(df['horsepower'])
plt.title('horsepower')
           
plt.subplot(1, 2, 2)
sns.boxplot(df['horsepower'])
plt.title('horsepower')

plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
sns.distplot(df['curb_weight'])
plt.title('Curb Weight')
           
plt.subplot(1, 2, 2)
sns.boxplot(df['curb_weight'])
plt.title('Curb Weight')

plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
sns.distplot(df['engine_size'])
plt.title('Engine Size')
           
plt.subplot(1, 2, 2)
sns.boxplot(df['engine_size'])
plt.title('Engine Size')

plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.jointplot(y="horsepower", x="curb_weight", data=df)
plt.title('Horsepower vs Weight', x=-2.7, y=1.2)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.jointplot(y="horsepower", x="engine_size", data=df)
plt.title('Horsepower vs Engine Size', x=-2.7, y=1.2)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.jointplot(y="curb_weight", x="engine_size", data=df)
plt.title('Curb Weight vs Engine Size', x=-2.7, y=1.2)
plt.show()
```

```python
plt.figure(figsize=(18,5))
sns.barplot(x="make",y="horsepower", data=df)
plt.title('Horsepower by Make')
plt.xticks(rotation=35)

plt.show()
```

```python
features = pd.get_dummies(df['make'])

features['usa'] = np.where((df['make'].isin(['chevrolet', 'dodge', 'mercury', 'plymouth'])), 1, 0)
features['japan'] = np.where((df['make'].isin(['honda', 'isuzu', 'mazda', 'mitsubishi', 'nissan', 'subaru', 'toyota'])), 1, 0)
features['italy'] = np.where((df['make'].isin(['alfa-romero'])), 1, 0)
features['germany'] = np.where((df['make'].isin(['audi', 'bmw', 'mercedes-benz', 'porsche','volkswagen'])), 1, 0)
features['england'] = np.where((df['make'].isin(['jaguar'])), 1, 0)
features['france'] = np.where((df['make'].isin(['peugot', 'renault'])), 1, 0)
features['sweden'] = np.where((df['make'].isin(['saab', 'volvo'])), 1, 0)
```

```python
features['horse_by_weight'] = df.horsepower / df.curb_weight
```

```python
corrmat = df.corr()

# Set up the matplotlib figure.
f, ax = plt.subplots(figsize=(12, 9))

# Draw the heatmap using seaborn
sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
```

```python
means = df[['curb_weight','engine_size','horsepower']].mean(axis=0)
stds = df[['curb_weight','engine_size','horsepower']].std(axis=0)
features['weight_enginesize_horse'] = ((df[['curb_weight','engine_size','horsepower']] - means) / stds).mean(axis=1)
```

```python
plotdf = df.loc[:, ['curb_weight','engine_size','horsepower']]
plotdf['weight_enginesize_horse'] = features['weight_enginesize_horse'] 
corrmat2 = plotdf.corr()

print(corrmat2)
```
