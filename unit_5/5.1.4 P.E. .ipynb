#### Problem 1 
If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
t = 0
for i in range(1000):
    if i % 3 == 0 or i % 5 == 0:
        t += i
print(t)
        
```

#### Problem 3 
The prime factors of 13195 are 5, 7, 13 and 29.

What is the largest prime factor of the number 600851475143 ?

```python
def large_prime_factor(number):
    i = 2
    while i * i < number:
        while number % i == 0:
            number = number / i
        i+= 1
    return number
    
```

#### Problem 5 
2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
# would like to discuss the while / for / break / else construct
result = 2520
done = False
while not done:
    result += 20
    for i in [20, 19, 18, 17, 16, 15, 14, 13, 11]:
        if result % i != 0:
            break
    else:
        done = True

print(result)
```

#### Problem 6 
The sum of the squares of the first ten natural numbers is,

1^2 + 2^2 + ... + 10^2 = 385
The square of the sum of the first ten natural numbers is,

(1 + 2 + ... + 10)^2 = 55^2 = 3025
Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 − 385 = 2640.

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```python
sum_of_squares = 0
numbers = list(range(1, 101))
for i in numbers:
    sum_of_squares += i**2


print(sum(numbers)**2 - sum_of_squares)    
```

#### Problem 10 
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
def is_prime(n):
    if n % 2 == 0:
        return False
    for i in np.arange(3, int(n**.5)+1, 2):
        if n % i == 0:
            return False
    return True

def prime_sum(number):
    total = 2
    for x in range(3, number, 2):
        if is_prime(x):
            total += x
    return total
```

```python
prime_sum(2000000)
```
