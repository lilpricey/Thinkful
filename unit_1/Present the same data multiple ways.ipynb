# Auto-mpg dataset
### Mileage per gallon performances of various cars

The Auto-mpg dataset contains the technical specifications of cars. It comes from the UCI Machine Learning Repository and is available [here](https://archive.ics.uci.edu/ml/datasets/Auto+MPG).

### The data contains the following columns:
* mpg: continuous
* cylinders: multi-valued discrete
* displacement: continuous
* horsepower: continuous
* weight: continuous
* acceleration: continuous
* model year: multi-valued discrete
* origin: multi-valued discrete
* car name: string (unique for each instance)

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
plt.rcParams['figure.figsize'] = 15, 10
```

## Miles Per Gallon

```python
plt.figure(figsize=(8,6))

sns.distplot(df['mpg'])
plt.title('Miles Per Gallon')

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(df['mpg'])
plt.title('Miles Per Gallon')

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.violinplot(df['mpg'])
plt.title('Miles Per Gallon')

plt.show()
```

## Horsepower and displacement 

```python
df = df[df['horsepower'] != '?']
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'])


plt.figure(figsize=(8,8))
sns.jointplot(y="horsepower", x="displacement", data=df)
plt.title('Horsepower vs Displacement', x=-2.7, y=1.2)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.jointplot(y="horsepower", x="displacement", kind="hex", data=df)
plt.title('Horsepower vs Displacement', x=-2.7, y=1.2)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.jointplot(y="horsepower", x="displacement", kind="kde", data=df)
plt.title('Horsepower vs Displacement', x=-2.7, y=1.2)
plt.show()
```

```python
plt.figure(figsize=(8,8))
cmap = sns.cubehelix_palette(as_cmap=True, rot=.75, dark=1, light=0, reverse=False)
sns.kdeplot(df["displacement"], df["horsepower"], cmap=cmap, n_levels=50, shade=True)
plt.title('Horsepower vs Displacement')
plt.show()
```

## Horsepower and Cylinders
